## Limpieza otros datos nulos

Son importantes los siguientes posibles features, y tienen datos nulos:

* rooms
* bedrooms
* bathrooms
* surface_total
* surface_covered
* property_type
* Ubicación: país-provincia-localidad

Se va imputar los valores faltantes de rooms, bedrooms, bathrooms con al mediana dentro de cada tipo de propiedad(property_type) porque las tipos de propiedades tienen caracteristicas diferentes.

```python
data_arg['property_type'].unique()
```

```python
data_arg['property_type'].value_counts()
```

Consideraciones especiales:

* Lote al ser un terreno, no tiene habitaciones, ni baños. Debería tener 0

* Una cochera tampoco. Se va reemplazar por 0

* Otro es muy ambiguo. Por eso la mediana podría ser engañosa, se decide eliminarla.


```python
# Asignar 0 a propiedades que no deberían tener ambientes
data_arg.loc[data_arg['property_type'].isin(['Lote', 'Cochera']), ['rooms', 'bedrooms', 'bathrooms']] = 0
```

```python
# eliminar tipo 'Otro' si querés un dataset más limpio
data_arg = data_arg[data_arg['property_type'] != 'Otro']
```

```python
# Imputación por mediana 
data_arg['rooms'] = data_arg.groupby('property_type')['rooms'].transform(lambda x: x.fillna(x.median()))
data_arg['bedrooms'] = data_arg.groupby('property_type')['bedrooms'].transform(lambda x: x.fillna(x.median()))
data_arg['bathrooms'] =data_arg.groupby('property_type')['bathrooms'].transform(lambda x: x.fillna(x.median()))
```

## Tratamiento outliers

### outliers: surface_total

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='property_type', y='surface_total', data=data_arg)
plt.title('Superficie total por tipo de propiedad (sin filtrar)')
plt.xticks(rotation=45)
plt.yscale('log')  # Si hay muchos valores extremos, esto lo aplana visualmente
plt.show()
```

1. Datos reales del mercado argentino
Los valores mínimos y máximos estimados vienen de observar miles de publicaciones reales


| Tipo de propiedad | Ejemplos reales                                          |
| ----------------- | -------------------------------------------------------- |
| Casa              | Desde 40 m² (casas pequeñas tipo PH) hasta >1000 m²      |
| Departamento      | Desde monoambientes (\~15 m²) hasta penthouses (>300 m²) |
| Lote              | Desde 100 m² en zona urbana hasta hectáreas (>10.000 m²) |
| Cochera           | Entre 12 y 50 m²                                         |
| Depósito          | 100 m² a decenas de miles, pero 10.000 es buen corte     |
| PH                | Similar a casas pequeñas: 30 m² a 500 m²                 |
| Oficina / Local   | Depende del rubro, pero suelen estar entre 20 y 2000 m²  |


```python
rango_surface = {
    'Casa': (30, 1500),
    'Departamento': (15, 500),
    'PH': (20, 600),
    'Lote': (100, 100000),  # más amplio
    'Cochera': (5, 100),
    'Oficina': (20, 3000),
    'Depósito': (50, 10000),
    'Local comercial': (20, 2000),
    'Casa de campo': (200, 20000)
}
```

```python
def filtrar_surface_por_tipo(df):
    df_filtrado = pd.DataFrame()
    for tipo, (min_val, max_val) in rango_surface.items():
        subset = df[df['property_type'] == tipo]
        subset = subset[(subset['surface_total'] >= min_val) & (subset['surface_total'] <= max_val)]
        df_filtrado = pd.concat([df_filtrado, subset])
    return df_filtrado
```

```python
# Paso 2: aplicás la función a tu dataset
data_filtrado = filtrar_surface_por_tipo(data_arg)
```

```python
data_filtrado.shape
```

```python
# Filtrado
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_filtrado, x='property_type', y='surface_total')
plt.title('Superficie total por tipo de propiedad (filtrado)')
plt.xticks(rotation=45)
plt.yscale('log')
plt.show()
```
